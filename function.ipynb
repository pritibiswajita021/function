** 6. Explain the concept of generators in Python and how they are defined.
**
* In Python, a generator is a special type of iterator that produces a sequence of values using the yield keyword.
* Generators are defined like regular functions, but instead of using the return statement to return a value, they use the yield.

** 7. What are the advantages of using generators over regular functions?
**
* Regular functions store all values in memory but Generators yield one value at a time – great for large data(memory efficiency).
* Generators start giving results immediately without building the whole list(faster execution).
* Values are generated on demand — ideal for streaming data, reading large files, etc.
* Generators automatically remember where they left off.

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square(a):
  n=[]
  for i in a:
    n.append(i**2)
  return n
square([1,2,3,4])

```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

prime_numbers = list(filter(is_prime, range(1, 201)))
print(prime_numbers)
```

```python
 #5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
 def fib(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a,b=b,a+b
for num in fib(15):
    print(num)

```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def square_generator(n):
  for i in range(n):
    yield 2**i
for num in square_generator(5):
    print(num)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
            for line in read_file('example.txt'):
                print(line)
```
